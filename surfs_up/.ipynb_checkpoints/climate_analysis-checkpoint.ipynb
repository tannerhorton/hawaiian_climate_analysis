# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session, load_only, sessionmaker
from sqlalchemy.sql import text, select
# code below is needed in order to connect to the data base, had to add inspect to starter code (D2A8)
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite, notice path must be included below 
# connection string tells us driver, dialect and File name 
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
#dir(Base.classes)
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link or connection) from Python to the DB
# engine is the common interface to the database from SQLAlchemy 
session = Session(engine)
# establish connection
conn = engine.connect()
```

```python
# Create the inspector and connect it to the engine, thise cell is pulled directly from day 2 Acitivity 8 
inspector = inspect(engine)
```

```python
# Collect the names of tables within the database
inspector.get_table_names()
# # Collect the names of tables within the database, better to use inspect function instead of code below 
# print(engine.table_names())
```

```python
# Using the inspector to print the column names within the 'Salaries' table and its types
columns = inspector.get_columns('measurement')
# remember you must iterate through because a list [] was returned in previous cell
for column in columns:
    print(column["name"], column["type"])
```

```python
# Using the inspector to print the column names within the 'Salaries' table and its types
columns = inspector.get_columns('station')
# remember you must iterate through because a list [] was returned
for column in columns:
    print(column["name"], column["type"])
```

# Exploratory Precipitation Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
measurements = session.query(measurement)
for measurement in measurements:
        print(measurement.date)
```

```python
# Find the most recent date in the data set.
most_recent_date = session.query(func.max(measurement.date)).one()
# session.query(measurement.date).nanmax() OR session.query(measurement).order_by(measurement.date.desc())
print(most_recent_date)
```

```python
# Starting from the most recent data point in the database. 
last_date = date(2017,8,23)
print(last_date)
```

```python
# Calculate the date one year from the last date in data set.
first_date = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(first_date)
```

```python
# Perform a query to retrieve the data and precipitation scores
# results = session.query(measurement.date, measurement.prcp).filter(measurement.date >= ('2016-08-23').all()
# st = pd.read_sql("SELECT * FROM measurement WHERE date BETWEEN '2016-8-23' AND '2017-8-23'",conn)
last_year = pd.read_sql_query("select Measurement.date, Measurement.prcp from measurement where date >= '2016-08-23'",conn)
last_year
```

```python
# Save the query results as a Pandas DataFrame 
type(last_year)
```

```python
# Sort the dataframe by date and set the index to the date column
ordered_dates = last_year.sort_values(by='date',ascending=True).dropna()
ordered_dates
```

```python
# Use Pandas Plotting with Matplotlib to plot the data, come back to change x axis to better match .png 
ordered_dates.reset_index().plot(x='date' , y= 'prcp',rot = 90)
plt.xlabel("date")
plt.ylabel("inches")
```

```python
#Use Pandas to print the summary statistics for the precipitation data
ordered_dates.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
total_stations = session.query(station).count()
print(total_stations)
Session = sessionmaker(bind = engine)
session = Session()
```

```python
# conn = engine.connect()
# saving station table as a data frame
station_df = pd.read_sql('select * from station', conn)
station_df.head()
```

```python
# saving measurment table as a dataframe
measurement_df = pd.read_sql('select * from measurement', conn)
measurement_df
```

```python
# check data type of desired column
type(measurement_df.station)
```

```python
# convert series to list in order to use counter function in next cell
station_list=measurement_df.station.tolist()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
data = Counter(station_list)
# List the stations and the counts in descending order.
data.most_common()   
```

```python
# Returns the highest occurring item
data.most_common(1)  
```

```python
# create data frame consisting of only temp values from the most active station
most_active = pd.read_sql("select measurement.tobs from measurement where measurement.station == 'USC00519281'", conn)
most_active
```

```python
# check data type
type(most_active)
```

```python
# convert column series to list 
temp_list=most_active.tobs.tolist()
temp_list
```

```python
# Using the most active station id from the previous query, calculate the lowest temp
lowest_temp = min(temp_list)
lowest_temp
```

```python
# Using the most active station id from the previous query, calculate the highest temp
highest_temp = max(temp_list)
highest_temp
```

```python
# Using the most active station id from the previous query, calculate the average temp
avg_temp = sum(temp_list)/len(temp_list)
avg_temp
```

```python
# Using the most active station query the last year of temperature data for this id & plot results as a histogram
most_active.hist(column='tobs', bins=12);
# come back to change x axis to better match .png 
plt.xlabel("Temperature")
plt.ylabel("Frequency")
```
